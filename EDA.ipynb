```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_pop=pd.read_excel('../input/ey-final/population census data.xlsx',header=None)
df_pop[1]=df_pop[1].apply(lambda x: x.upper())
df_pop
```

```python
pop_dict=dict(zip(df_pop.iloc[:,1],df_pop.iloc[:,2]))
pop_dict
```

```python
def Population_plot():
    fig_dims = (15, 8)
    fig, ax = plt.subplots(figsize=fig_dims)
    sns.barplot(df_pop.iloc[:,1],df_pop.iloc[:,2])
    ax.set_xlabel("State")
    ax.set_ylabel('Population')
    ax.set_title('Population Across Different States in India')
    plt.ticklabel_format(style='sci', axis='y', scilimits=(1,10))
    plt.xticks(rotation=90)
    plt.show()
    
```

# Population distribution of india

```python
df_death=pd.read_csv('../input/ey-final/Monthly deaths across states (1).csv')
df_positive=pd.read_csv('../input/ey-final/Monthly positive across states.csv')
df_death.drop(columns=['UN'],inplace=True)
df_positive.drop(columns=['UN'],inplace=True)
```

```python
df_death.head()
```

```python
df_positive.head()
```

```python
df_cumulative=pd.DataFrame()
df_cumulative["Death"]=df_death.iloc[:,2:].sum()
df_cumulative["positive"]=df_positive.iloc[:,2:].sum()
df_cumulative=df_cumulative.T
df_cumulative.to_csv("Cumulative Death and Positive.csv")
```

```python
df_cumulative
```

```python
dates=['March','April','May','June','July','August','September','October','November','December','Jan-2021']
```

# Death_Trend_Across_India_lineplot

```python
def Death_Trend_Across_India_lineplot(): 
    fig_dims = (10, 5)
    fig, ax = plt.subplots(figsize=fig_dims)
    plt.plot(dates,df_death['TT'])
    ax.set_xlabel("Month")
    ax.set_ylabel('Deaths')
    ax.set_title('Deaths from COVID19 in India (March 2020- Januray 2021')
    plt.xticks(rotation=90)
    plt.show()
```

# Death_Trend_Across_India_barplot

```python
def Death_Trend_Across_India_barplot():
    fig_dims = (10, 5)
    fig, ax = plt.subplots(figsize=fig_dims)
    k=sns.barplot(x=dates,y=df_death['TT'])
    plt.xticks(rotation=90)
    k.set(xlabel='Month',ylabel='deaths')
    k.set(title='Deaths from COVID19 in India (March 2020- Januray 2021)')
    plt.show()
```

# Positive_Cases_Across_India_lineplot

```python
def Positive_Cases_Across_India_lineplot(): 
    fig_dims = (10, 5)
    fig, ax = plt.subplots(figsize=fig_dims)
    plt.plot(dates,df_positive['TT'])
    ax.set_xlabel("Month")
    ax.set_ylabel('Deaths')
    ax.set_title('Deaths from COVID19 in India (March 2020- Januray 2021')
    plt.ticklabel_format(style='sci', axis='y', scilimits=(1,10))
    plt.xticks(rotation=90)
    plt.show()
```

# Positive_Cases_Across_India_barplot

```python
def Positive_Cases_Across_India_barplot():
    fig_dims = (10, 5)
    fig, ax = plt.subplots(figsize=fig_dims)
    k=sns.barplot(x=dates,y=np.array(df_positive['TT']))
    plt.xticks(rotation=90)
    plt.ticklabel_format(style='sci', axis='y', scilimits=(1,10))
    k.set(xlabel='Month',ylabel='Cases')
    k.set(title='Positive Cases from COVID19 in India (March 2020- Januray 2021)')
    plt.show()
```

# State_Wise_Positive_Cases_Analysis_linplot 

```python
def State_Wise_Positive_Cases_Analysis_linplot(state):
    state=state.upper()
    code=state_code[state]
    fig_dims = (10, 5)
    fig, ax = plt.subplots(figsize=fig_dims)
    
    plt.plot(dates,np.array(df_positive[code]))
    plt.xticks(rotation=90)
    plt.ticklabel_format(style='sci', axis='y', scilimits=(1,10))
    ax.set_xlabel("Month")
    ax.set_ylabel('+ve')
    ax.set_title('+ve cases of COVID19 in '+state+ ' (March 2020- Januray 2021')
    plt.show()
    
    
```

```python
df_statecode=pd.read_excel('../input/ey-final/state code.xlsx')
df_statecode=df_statecode.drop(axis=0,index=2)
df_statecode['Name of the state']=df_statecode['Name of the state'].apply(lambda x: x.upper())
```

```python
kkk=list(df_positive.columns)[2:]
sts=list(df_statecode['Name of the state'])
state_code=dict(zip(sts,kkk))
```

# State_Wise_Positive_Cases_Analysis

```python
def State_Wise_Positive_Cases_Analysis(state):
    state=state.upper()
    code=state_code[state]
    fig_dims = (10, 5)
    fig, ax = plt.subplots(figsize=fig_dims)
    k=sns.barplot(x=dates,y=np.array(df_positive[code]))
    plt.xticks(rotation=90)
    plt.ticklabel_format(style='sci', axis='y', scilimits=(1,10))
    k.set(xlabel='Month',ylabel='Cases')
    k.set(title="Positive Cases from COVID19 in "+str(state.capitalize())+" (March 2020 - Januray 2021)")
    plt.show()
    
    
```

# State_Wise_Death_Analysis

```python
def State_Wise_Death_Analysis(state):
    state=state.upper()
    code=state_code[state]
    fig_dims = (10, 5)
    fig, ax = plt.subplots(figsize=fig_dims)
    k=sns.barplot(x=dates,y=np.array(df_death[code]))
    plt.xticks(rotation=90)
    plt.ticklabel_format(style='sci', axis='y', scilimits=(0,10))
    k.set(xlabel='Month',ylabel='Cases')
    k.set(title="Number of Deaths due to COVID19 in "+str(state.capitalize())+" (March 2020 - Januray 2021)")
    plt.show()
    
```

```python
def Population_vs_positive_cases_StateWise(state):
    state=state.upper()
    code=state_code[state]
    lst=([pop_dict[state],df_cumulative[code][1],df_cumulative[code][0]])
    labels=["Population","Positive","Deaths"]
    fig_dims = (8, 10)
    fig, ax = plt.subplots(figsize=fig_dims)
    k=sns.barplot(x=labels,y=lst)
    plt.ticklabel_format(style='sci', axis='y', scilimits=(0,10))
    k.set(ylabel='Population')
    k.set(title="Popolation vs Positive vs Death in "+str(state.capitalize())+" (March 2020 - Januray 2021)")
    plt.show()
    
```

```python
my_dict={1:Population_plot,
        2:Death_Trend_Across_India_lineplot,
        3:Death_Trend_Across_India_barplot,
        4:State_Wise_Positive_Cases_Analysis_linplot,
        5:State_Wise_Positive_Cases_Analysis,
        6:State_Wise_Death_Analysis,
        7:Population_vs_positive_cases_StateWise}
```
